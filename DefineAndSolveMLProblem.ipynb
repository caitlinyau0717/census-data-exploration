Address missingness, replace outliers, replace the name of fnlwgt, one-hot encoding of gender, race, workclass, education, occupation, relationship

```python
# THIS IS FOR MISSING VALUES:

# Assuming df is your DataFrame containing the dataset
# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)

# Replace missing numerical values with mean
df['age'].fillna(df['age'].mean(), inplace=True)
df['hours-per-week'].fillna(df['hours-per-week'].mean(), inplace=True)

# Replace missing categorical values with mode
df['workclass'].fillna(df['workclass'].mode()[0], inplace=True)

# Example: Drop rows with missing values in specific columns
df.dropna(subset=['occupation', 'native-country'], inplace=True)

# Verify if missing values have been handled
print("After handling missing values:")
print(df.isnull().sum())
```

```python
# THIS IS FOR OUTLIERS

import scipy.stats as stats

hpw_999 = np.percentile(df['hours-per-week'], 99.9)

df['hours-per-week-win'] = stats.mstats.winsorize(df['hours-per-week'], limits=[0.01, 0.01])
df.drop(columns=['hours-per-week'], inplace=True)
df.head()
```

```python
# THIS IS FOR REPLACING NAMES:

columnDict = {'fnlwgt': 'final-weight'}

df = df.rename(columns=columnDict)
df.head()
```

```python
# THIS IS FOR ONE-HOT ENCODING

to_encode = list(df.select_dtypes(include=['object']).columns)
to_encode.remove('income_binary')

df[to_encode].nunique()

df_workclass = pd.get_dummies(df['workclass'], prefix='workclass_')
df = df.join(df_workclass)
df.drop(columns = 'workclass', inplace=True)

df_education = pd.get_dummies(df['education'], prefix='education_')
df = df.join(df_education)
df.drop(columns = 'education', inplace=True)

df_marital_status = pd.get_dummies(df['marital-status'], prefix='marital-status_')
df = df.join(df_marital_status)
df.drop(columns = 'marital-status', inplace=True)

df_occupation = pd.get_dummies(df['occupation'], prefix='occupation_')
df = df.join(df_occupation)
df.drop(columns = 'occupation', inplace=True)

df_relationship = pd.get_dummies(df['relationship'], prefix='relationship_')
df = df.join(df_relationship)
df.drop(columns = 'relationship', inplace=True)

df_race = pd.get_dummies(df['race'], prefix='race_')
df = df.join(df_race)
df.drop(columns = 'race', inplace=True)

df_sex_selfID = pd.get_dummies(df['sex_selfID'], prefix='sex_')
df = df.join(df_sex_selfID)
df.drop(columns = 'sex_selfID', inplace=True)

df_country = pd.get_dummies(df['native-country'], prefix='country_')
df = df.join(df_country)
df.drop(columns = 'native-country', inplace=True)

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

<b>Task:</b> Use the rest of this notebook to carry out your project plan. 

You will:

1. Prepare your data for your model.
2. Fit your model to the training data and evaluate your model.
3. Improve your model's performance by performing model selection and/or feature selection techniques to find best model for your problem.

Add code cells below and populate the notebook with commentary, code, analyses, results, and figures as you see fit. 

```python
# Sorry i already prepared the data above

# setting up the data
X = df.drop('income_binary', axis=1)
y = df['income_binary']

#splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1234)

# standardizing the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# fitting the data to the model
model = LogisticRegression(random_state=42)
model.fit(X_train_scaled, y_train)

# analyzing the model's performance
y_pred = model.predict(X_test_scaled)
print("Accuracy:", accuracy_score(y_test, y_pred))
c_m = confusion_matrix(y_test, y_pred)
pd.DataFrame(
c_m,
columns=['Predicted: Income <= 50k', 'Predicted: Income > 50k'],
index=['Actual: Income <= 50k', 'Actual: Income > 50k']
)
```

Looking at these results, my model performed pretty well in my opinion as the accuracy score of 84.79% is pretty good and as seen by the confusion matrix, the incorrect predictions are pretty small compared to the actual prediction numbers.
